**Question 7: What is the significance of confidence intervals in statistical analysis?**

Ans7 *A confidence interval is a range of values, derived from sample data, that is likely to contain the true population parameter (like mean or proportion) with a specified confidence level (e.g., 95%).*

**Importance:**

**Estimation of population parameters:**

Instead of giving a single “point estimate” (like a sample mean), CI provides a range, which accounts for sampling variability.

**Quantifying uncertainty:**

CI tells us how precise our estimate is. A narrower CI → more precise; a wider CI → less precise.

**Decision-making**:

Used in hypothesis testing, quality control, and business/medical studies to assess reliability of results.

***Generalization:***

Helps infer population characteristics from sample data.

```python
'''Question 9: Write a Python program to generate 1000 random numbers from a normal
distribution with mean = 50 and standard deviation = 5. Compute its mean and standard
deviation using NumPy, and draw a histogram to visualize the distribution.
(Include your Python code and output in the code box below.) '''


import numpy as np
import matplotlib.pyplot as plt


mean = 50
std_dev = 5
n = 1000

# Generate 1000 random numbers from normal distribution
data = np.random.normal(loc=mean, scale=std_dev, size=n)

# Compute mean and standard deviation
computed_mean = np.mean(data)
computed_std = np.std(data)

print("Computed Mean:", computed_mean)
print("Computed Standard Deviation:", computed_std)

# Plot histogram
plt.hist(data, bins=30, color='skyblue', edgecolor='black')
plt.title("Histogram of Normal Distribution (mean=50, std=5)")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()

```

```python
'''Question 10: You are working as a data analyst for a retail company. The company has
collected daily sales data for 2 years and wants you to identify the overall sales trend.
daily_sales = [220, 245, 210, 265, 230, 250, 260, 275, 240, 255,
 235, 260, 245, 250, 225, 270, 265, 255, 250, 260]
● Explain how you would apply the Central Limit Theorem to estimate the average sales
with a 95% confidence interval.
● Write the Python code to compute the mean sales and its confidence interval.
(Include your Python code and output in the code box below.) '''

# Ans 9

# CLT states that the sampling distribution of the sample mean will be approximately normal if the sample size is sufficiently large (n ≥ 30 is typical, but smaller sizes can work with roughly symmetric distributions).

# Using CLT, we can estimate the average daily sales and compute a 95% confidence interval

import numpy as np
from scipy import stats

# Daily sales data
daily_sales = [220, 245, 210, 265, 230, 250, 260, 275, 240, 255,
               235, 260, 245, 250, 225, 270, 265, 255, 250, 260]

# Sample mean and standard deviation
mean_sales = np.mean(daily_sales)
std_sales = np.std(daily_sales, ddof=1)  # ddof=1 for sample std
n = len(daily_sales)

# 95% Confidence Interval
confidence = 0.95
z_score = stats.norm.ppf(0.975)  # two-tailed 95% CI

margin_of_error = z_score * (std_sales / np.sqrt(n))
ci_lower = mean_sales - margin_of_error
ci_upper = mean_sales + margin_of_error

print("Mean Daily Sales:", mean_sales)
print("95% Confidence Interval: ({:.2f}, {:.2f})".format(ci_lower, ci_upper))
```
